```python
# ------ Main Class ------
def Scheduling(Cwl, Rm ):
    for i in range(3):      # ให้ตารางกลับมาเป็น xxxxxxx
        for j in range(9):
            Rm[i,j] = ["xxxxxxx"]

    
    for i in range(len(Cwl)):
        rdRoom = random.randint(0,len(Rm))
        rdSlot = random.randint(0,len(Rm[0]))
        while (Rm[rdRoom, rdSlot] != "xxxxxxx"):
            rdRoom = random.randint(0,len(Rm))
            rdSlot = random.randint(0,len(Rm[0]))
        Cwl[i,2] = rdRoom    #Cwl[i,2] = random.randint(0,len(Rm)-1)
        Cwl[i,3] = rdSlot    #Cwl[i,3] = random.randint(0,len(Rm[0])-1)
        Rm[int(Cwl[i,2]), int(Cwl[i,3])] = Cwl[i,0]  #Rm[int(Cwl[i,2]), int(Cwl[i,3])] = Cwl[i,0]
    

def ComputeFittnessValue(Rm , cwl ,mp ):
    s=[]          # เก็บข้อมูลทีม 
    value = []    # เก็บจำนวนงานรวมที่แต่ละทีมทำ
    t=[]          # เปลี่ยนเซตให้เป็นเป็นอาเร
    for i in mp:
        t.append(i)
        
    for i in range(len(cwl)): # เช็ตงานของทีมพนักงานว่าซ้อนกันไหม
        s.append(cwl[i,2])
        for j in range(len(cwl)):
            if i != j:
                if (cwl[i,3] == cwl[j,3]):
                     if (cwl[i,2] == cwl[j,2]):
                        print ("พนักงานมีงานซ้อนกัน")
                        
    print ("---------------\n","ภาระงานทั้งหมดมี",len(cwl)) #for loop เพื่อเก็บจำนวนงานรวมที่แต่ละทีมทำ
    for i in range(len(mp)):
        value.append(0)
        for j in range(len(cwl)):
            if i == int(s[j]):
                value[i] = value[i]+1 
    

    for i in range(-1,len(value)-1):  # จัดให้เป็นตารางทีมกับงานที่ทีมพนักงานได้รับทั้งหมดเพื่อให้ดูง่าย
        print ("พนักงานทีม :",t[i+1],value[i])
    
    return value
 
    
def DisplayRoom(Rm):
    for i in range(len(Rm)):
        print()
        for sl in range(len(Rm[i])):
            print(Rm[i,sl], end=" ")
            if sl%2 == 1 :
                print()

def DisplayCourseWorkLoad(cwl):
    print('\n',cwl)

# np.random.seed(0) 
T = Work()
R = Team()
C = CoursesWorkload()
    
while (True): # เมื่อมีงานที่ไม่เท่าเทียมกันจะ มีการ random ใหม่จดกว่าจะได้งานที่เท่าเทียมกัน
    Scheduling(C.getCourseWorkload(), R.getTeamWeek())
    ch = ComputeFittnessValue(R.getTeamWeek(),C.getCourseWorkload(),R.getTeamName())
    if  min(ch) == max(ch) or min(ch) == max(ch)-1:
        print ("||เป็นภาระงานที่เท่าเทียมกัน||")
        break

print ("\n**ตารางภาระงานที่เท่าเทียมกัน**","\n_รหัสงาน เลขงาน ทีม เวลา_")
DisplayCourseWorkLoad(C.getCourseWorkload())
DisplayRoom(R.getTeamWeek())

```
