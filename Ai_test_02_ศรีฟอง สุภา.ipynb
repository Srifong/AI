```python
import numpy as np
from numpy import random


def City(st):
    for i in range(len(cities)):
        if st == cities[i]:
            return i

def Random(ad, st, t, num): # random แต่ครั้ง
    ran = 0
    for i in ad[st]:
        t = t + 1
        if i != 0 :
            num.append(t)
    ran = num[random.randint(0, len(num))]
    return ran


def check(cities, ad, need, st, sp, reslut): # ฟังก์ชั่นเช็ค
    sum = 0
    sumneed = []
    Set = []
    for _ in range(1000): # สามารถเปลี่ยนเลขเมื่อให้แม่นยำขึ้น
        S = []
        I = st
        i = 0
        while sp != I:
            i = Random(ad,I,-1,[])
            sum = sum + ad[I,i]
            S.append(cities[i])
            I = i
        reslut.append(sum)
        Set.append(S)
        sum = 0
        I = st

    for i in reslut: # เช็คว่าระยะทางที่รับเข้ามาว่าใกล้เคียงกับที่ random ให้มากที่สุด
        if i > int(need):
            sumneed.append(i - int(need))
        else:
            sumneed.append(int(need) - i)
    # print(Set)
    # print(reslut)
    # print(sumneed)
    for i in range(len(sumneed)): # จัดตัวหนังสือ
        if min(sumneed) == sumneed[i]:
            print("เส้นทางที่ดีที่สุดคือจาก :",cities[st],"ไป",Set[i],"ระยะทางเท่ากับ",reslut[i])


cities = ["A", "B", "C", "D", "E"]
adjacency_mat = np.asarray([
    [0, 2, 5, 8, 0],
    [2, 0, 0, 3, 4],
    [5, 0, 0, 4, 3],
    [8, 3, 4, 0, 2],
    [0, 4, 3, 2, 0],
])

need = input("ระบุระยะทางที่ต้องการ :")
st = City(input("จุดเริ่มต้น :"))
sp = City(input("จุดหมายปลายทาง :"))
sum = 0
print("-------------")
check(cities, adjacency_mat, need, st, sp, [])
```
